```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report 
```

```python
features=['WTT', 'PTI', 'EQW', 'SBI', 'LQE', 'QWG', 'FDJ', 'PJF', 'HQE', 'NXJ']
X = knn[features]
Y = knn['TARGET CLASS']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size = 0.20,random_state=101)
```

```python
knn_model = KNeighborsClassifier()
knn_model.fit(X_train,Y_train)#n_neighbors=5 is default if we not mention any n_neighbours then it will take 5 automatically
```

```python
Y_predictedtest = knn_model.predict(X_test)
#Y_predictedtrain=knn_model.predict(Y_test)
```

```python
Y_predictedtest
```

```python
print(accuracy_score(Y_test,Y_predictedtest))
```

# Grid SearchCv

Grid search builds a model for every combination(permutations and combination) of hyperparameters specified and evaluates each model. 

```python
from sklearn.model_selection import GridSearchCV
params = {'n_neighbors':list(np.arange(2,21))}
```

```python
knn_cv = KNeighborsClassifier()

model = GridSearchCV(knn_cv,params,cv=5)
```

```python
import time
start = time.time()
model.fit(X_train1,Y_train1)
end = time.time()
training_time = end-start
print("Time taken:",training_time)
```

```python
knn_10 = KNeighborsClassifier(n_neighbors=10)
knn_10.fit(X_train1,Y_train1)
pred_10=knn_10.predict(X_test1)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(accuracy_score(Y_test1,pred_10))
print(classification_report(Y_test1,pred_10))
print(confusion_matrix(Y_test1,pred_10))
```

```python
error_rate = []
for i in range(1,21):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,Y_train)
    predictions_i= knn.predict(X_test)
    error_rate.append(np.mean(predictions_i != Y_test))
    
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))
plt.plot(range(1,21),error_rate,color='blue',linestyle = '--',marker ='o',markerfacecolor = 'aqua',markersize = 10)

plt.title('Error _Rate vs K_vale')
plt.xlabel("k_value")
plt.ylabel("Error rate")
plt.show();
```

```python
Accuracy = []
for i in range(1,21):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,Y_train)
    predictions_i= knn.predict(X_test)
    Accuracy.append(np.mean(predictions_i == Y_test))
```

```python
plt.figure(figsize=(20,10))
plt.plot(range(1,21),Accuracy,color='blue',linestyle = '--',marker ='o',markerfacecolor = 'aqua',markersize = 10)

plt.title('Accuracy vs K_vale')
plt.xlabel("k_value")
plt.ylabel("Accuracy")
plt.show();
```
